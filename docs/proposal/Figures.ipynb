```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
R = 1
M = 1
ssb = np.arange(0, M * 20, 0.01)
r = R * ssb / (1 + ssb / M)

df = pd.DataFrame({'Spawners': ssb, 'Recruits': r})

fig = px.line(df, x='Spawners', y='Recruits')
fig.update_layout(xaxis=dict(showgrid=False),
              yaxis=dict(showgrid=False)
)
fig.update_layout(
    plot_bgcolor='white'
)
fig.update_traces(line_color='black')
fig.show()
```

```python
with open('recruitment.png', 'wb') as fh:
    fh.write(fig.to_image("png"))
```
